```r
library ('tidyverse')
```

```r
turbine_data = read_csv ("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-10-27/wind-turbine.csv")
```

```r
turbine_data |> head()
```

Each row in the table represents a single turbine.

Make a bar chart ðŸ“Š **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
turbine_sorted = turbine_data |>
    mutate (province_sorted = province_territory |> fct_infreq())

p = ggplot (turbine_sorted, aes(x=province_sorted)) + geom_bar() + labs(title= 'Turbine Counts Per Province Territory')+
labs(x = 'Province Territory', y = 'Turbine Count') + coord_flip() 
p
```

Make a bar chart **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
turbine_new = turbine_data |>
    mutate (manufacturer = manufacturer |> fct_lump (n=5) |> fct_infreq() |> fct_rev())

p = ggplot(turbine_new, aes(x=manufacturer)) + geom_bar() + labs(title = 'Turbine Count Per Manufacturer') +
labs (x= 'Manufacturer', y='Turbine Count') + coord_flip()
p
```
