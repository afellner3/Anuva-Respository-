Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
options(repr.plot.height = 3, repr.plot.width = 4)

p = ggplot(penguins, aes(x = body_mass_g, color = species, fill = species)) + 
    geom_density(alpha = 0.25) + geom_rug() +
    scale_color_brewer(palette = 'Dark2') +
    scale_fill_brewer(palette = 'Dark2') + 
    theme_minimal() + 
    theme(legend.position = 'top') +
    labs(x = 'Body mass in grams')
p
```

```r
p = ggplot(penguins, aes(x=body_mass_g))
p = p + geom_histogram(fill='violet', color='steelblue')
p = p + theme_minimal() 
p = p + facet_wrap(~year, ncol=1, strip.position="right") + theme(strip.text = element_text(face = "bold"))
p = p + labs(x='Body mass in grams', y='Number of penguins', title="Histograms of penguin body mass for measurements taken from 2007-2009")
p = p + scale_x_continuous(labels=comma) + theme(title=element_text(size=8))
p = p + scale_y_continuous(breaks=c(0.0,2.5,5,7.5))
p
```

Is there a difference between the body mass of male and female penguins? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to compare within each species-group.**

```r
p = ggplot(penguins, aes(x=body_mass_g, y=sex))
p = p + geom_point(position=position_jitter(w=0.5, h=0.3), alpha=0.5)
p = p + theme_minimal() 
p = p + facet_wrap(~species, ncol=1) + theme(strip.text = element_text(face = "bold"))
p = p + labs(x="Body mass in grams") + scale_x_continuous(labels=comma, breaks=c(3500,4500,5500)) + theme(axis.title.y=element_text(color="white"))
p 
```
